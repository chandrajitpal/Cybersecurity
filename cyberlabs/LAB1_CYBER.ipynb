**The types of machine learning are:**



1. Supervised Learning

Definition: In supervised learning, the algorithm is trained using an input dataset where the desired output is already known. The primary goal is for the algorithm to learn the relationships between the input data and the provided target variables, often called "labels."

Application in Cybersecurity: A classic example is a spam filter. The algorithm is trained with a large number of emails that have been pre-classified as either "spam" (malicious/unwanted) or "ham" (genuine/harmless). Based on this training, the filter learns to classify new, incoming emails.

Common Algorithms:

    Regression (Linear and Logistic)

    k-Nearest Neighbors (k-NNs)

    Support Vector Machines (SVMs)

    Decision Trees and Random Forests

    Neural Networks (NNs)

2. Unsupervised Learning

Definition: In unsupervised learning, algorithms are tasked with classifying data independently, without the help of a pre-classified dataset. The model must identify patterns, structures, and relationships within the data on its own.

Application in Cybersecurity: This approach is crucial for identifying new and previously unseen threats. It is particularly important for detecting novel forms of malware, new types of financial fraud, and emerging email spamming campaigns that don't match existing signatures.

Common Algorithms:

    Dimensionality Reduction:

    Principal Component Analysis (PCA)

    PCA Kernel

    Clustering:

    k-means

    Hierarchical Cluster Analysis (HCA)

3. Reinforcement Learning (RL)

Definition: Reinforcement learning employs a "trial and error" approach. The algorithm, or "agent," learns by interacting with an environment. It receives positive rewards for correct decisions and negative rewards (or punishments) for incorrect ones, with the ultimate goal of maximizing its final reward.

Application in Cybersecurity: A key application is the detection of polymorphic malware. Algorithms like Hidden Markov Models (HMMs), which are based on the Markov process, are extremely important for identifying threats that constantly change their form to evade detection.

Common Algorithms:

    Markov Process

    Q-learning

    Temporal Difference (TD) methods

    Monte Carlo methods







**Supervised learning example  linear regression**

As our first example, we'll look at one of the most commonly used algorithms in the field of supervised learning, namely linear regression. Taking advantage of the scikit-learn Python library, we instantiate a linear regression object, by importing the LinearRegression class included in the linear_model package of the scikit-learn library.

The model will be trained with a training dataset obtained by invoking the rand() method of the RandomState class, which belongs to the random package of the Python numpy library. The training data is distributed following the linear model of, Y=3X+2. The training of the model is carried out by invoking the fit() method on the lreg object of the LinearRegression class. At this point, we will try to predict data that is not included in the training dataset by invoking the predict() method on the lreg object. The training dataset, together with the values interpolated by the model, are finally printed on screen using the scatter() and plot() methods of the matplotlib library:

```


%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

pool = np.random.RandomState(10)
x = 5 * pool.rand(30)
y = 3 * x - 2 + pool.randn(30)

# y = 3x - 2;
lregr = LinearRegression(fit_intercept=False)

X = x[:, np.newaxis]
lregr.fit(X, y)

lspace = np.linspace(0, 5)
X_regr = lspace[:, np.newaxis]
y_regr = lregr.predict(X_regr)

plt.scatter(x, y);
plt.plot(X_regr, y_regr);




# The preceding code generates the following output, which shows how well the data
# samples are approximated by the straight line returned by the LinearRegression model:
```

***Code Explanation: ***

The selected code demonstrates a simple linear regression example using the scikit-learn library as explained below:

    %matplotlib inline: This is a magic command in Jupyter/IPython environments (like Colab) that ensures that plots generated by matplotlib are displayed directly within the notebook.

    import matplotlib.pyplot as plt, import numpy as np, from sklearn.linear_model import LinearRegression: These lines import the necessary libraries: matplotlib.pyplot for plotting, numpy for numerical operations, and LinearRegression from scikit-learn for performing linear regression.

    pool = np.random.RandomState(10): This creates a RandomState object with a seed of 10. Using a fixed seed ensures that the random numbers generated are the same each time the code is run, making the results reproducible.

    x = 5 * pool.rand(30): This generates an array x of 30 random numbers between 0 and 1 using pool.rand(30) and then scales them by multiplying by 5, so the values are between 0 and 5. These will be our independent variable data points.

    y = 3 * x - 2 + pool.randn(30): This generates the dependent variable data y. It's created based on a linear relationship with x (3 * x - 2), and then adds some random noise using pool.randn(30), which generates random numbers from a standard normal distribution. This simulates real-world data where there's a general linear trend but also some variability.

    lregr = LinearRegression(fit_intercept=False): This initializes a LinearRegression model object. fit_intercept=False means the model will assume the line passes through the origin (0,0) and will not calculate an intercept.

    X = x[:, np.newaxis]: This reshapes the x array into a 2D array. scikit-learn models typically expect the input features (X) to be in a 2D format, where each row is a sample and each column is a feature. np.newaxis adds a new dimension.

    lregr.fit(X, y): This is the core of the linear regression. It trains the LinearRegression model using the prepared feature data X and the target variable y. The model learns the best-fit line that minimizes the difference between the predicted and actual y values.

    lspace = np.linspace(0, 5): This creates an array lspace of evenly spaced numbers between 0 and 5. This will be used to generate points for plotting the regression line smoothly across the range of our x data.

    X_regr = lspace[:, np.newaxis]: Similar to step 7, this reshapes the lspace array into a 2D array, preparing it for prediction by the scikit-learn model.

    y_regr = lregr.predict(X_regr): This uses the trained lregr model to predict the y values for the points in X_regr. These predicted values will form the linear regression line.

    plt.scatter(x, y): This creates a scatter plot of the original data points (x and y).

    plt.plot(X_regr, y_regr): This plots the predicted linear regression line using the X_regr points and their corresponding y_regr predictions.

This code generates some sample data that roughly follows a linear pattern with noise, trains a linear regression model on this data, and then plots the original data points and the learned regression line to visualize the relationship.

**Unsupervised Learning Example: Clustering**

As an example of unsupervised learning, we use the GaussianMixture clustering model.

Through this type of model, we will try to bring the data back to a collection of Gaussian

Blobs. The training data is loaded from a file in .csv format (comma-separated values) and stored in a DataFrame object of the pandas Python library. Once the data is loaded, we proceed to reduce its dimensionality in order to identify a representation that reduces the original dimensions (features) from four to two, trying to maintain the features that are most representative of the samples.

The reduction of dimensionality prevents the disadvantages connected to the phenomenon of the curse of dimensionality, improves the computational efficiency, and simplifies the visualization of the data.

The technique we will use for dimensionality reduction is known as principal component

analysis (PCA), and is available in the scikit-learn library. Once the data dimensions are reduced from four to two, we will try to classify the data using the GaussianMixture model as follows:


```
import pandas as pd
import seaborn as sns


from google.colab import drive
drive.mount('/content/drive')





file_path = '/content/drive/My Drive/PICTURES_CYBER/clustering.csv'

data_df = pd.read_csv(file_path)
data_df.describe()

X_data = data_df.drop('class_1', axis=1)
y_data = data_df['class_1']

from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(X_data)
X_2D = pca.transform(X_data)
data_df['PCA1'] = X_2D[:, 0]
data_df['PCA2'] = X_2D[:, 1]

from sklearn.mixture import GaussianMixture

gm = GaussianMixture(n_components=3, covariance_type='full')
gm.fit(X_data)

y_gm = gm.predict(X_data)
data_df['cluster'] = y_gm

sns.lmplot(x="PCA1", y="PCA2", data=data_df, col='cluster', fit_reg=False)
```

***Code explanation: ***

The selected code cell performs unsupervised learning, specifically clustering, on your data using Gaussian Mixture Models (GMM) as shown below:

    Import Libraries: It imports necessary libraries: pandas for data manipulation, seaborn for visualization, matplotlib.pyplot for plotting, numpy for numerical operations, and modules from sklearn for PCA and Gaussian Mixture Models.

    Mount Google Drive: It mounts your Google Drive to access the clustering.csv file.

    Load Data: It reads the clustering.csv file into a pandas DataFrame called data_df.

    Data Preparation:

    data_df.describe(): Provides descriptive statistics of the data.

    X_data = data_df.drop('class_1', axis=1): Creates a new DataFrame X_data by dropping the column named 'class_1'. This assumes 'class_1' is a label or target variable and the remaining columns are features for clustering.

    y_data = data_df['class_1']: Creates a Series y_data containing the 'class_1' column.

    Dimensionality Reduction (PCA):

    pca = PCA(n_components=2): Initializes a PCA object to reduce the data to 2 principal components.

    pca.fit(X_data): Fits the PCA model to your feature data (X_data).

    X_2D = pca.transform(X_data): Transforms the data into the 2-dimensional PCA space.

    data_df['PCA1'] = X_2D[:, 0] and data_df['PCA2'] = X_2D[:, 1]: Adds the two principal components as new columns ('PCA1' and 'PCA2') to the original DataFrame.

    Clustering (Gaussian Mixture Model):

    gm = GaussianMixture(n_components=3, covariance_type='full'): Initializes a Gaussian Mixture Model object. n_components=3 specifies that the model should find 3 clusters, and covariance_type='full' specifies the type of covariance matrix to use.

    gm.fit(X_data): Fits the GMM model to the feature data.

    y_gm = gm.predict(X_data): Predicts the cluster assignments for each data point.

    data_df['cluster'] = y_gm: Adds the predicted cluster assignments as a new column ('cluster') to the DataFrame.

    Visualization:

    sns.lmplot(x="PCA1", y="PCA2", data=data_df, col='cluster', fit_reg=False): Creates a scatter plot using seaborn's lmplot.

    x="PCA1" and y="PCA2": Specifies the columns to use for the x and y axes (the principal components).

    data=data_df: Specifies the DataFrame to use.

    col='cluster': Creates separate plots for each cluster.

    fit_reg=False: Turns off the linear regression line.

The code takes the input data, reduces its dimensions using PCA, applies a Gaussian Mixture Model to group the data into 3 clusters, and then visualizes these clusters in the 2D PCA space.

**Simple Neural Network Example: Perceptron**


In this section, we will show a simple NN model, known as a perceptron. NNs and DL are subfields of ML aimed at emulating the human brain's learning
capabilities. However rudimentary it is, a perceptron is nonetheless able to adequately classify samples that tend to group together (in technical terms, those that are linearly separable). One of the most common uses of a perceptron in the field of cybersecurity, as we will see, is in the area of spam filtering.
In the following example, we will use the scikit-learn implementation of the perceptron algorithm:


```


from matplotlib.colors import ListedColormap
import numpy as np
import matplotlib.pyplot as plt



# Thanks to Sebastian Raschka for 'plot_decision_regions' function

def plot_decision_regions(X, y, classifier, resolution=0.02):

# setup marker generator and color map

  markers = ('s', 'x', 'o', '^', 'v')
  colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
  cmap = ListedColormap(colors[:len(np.unique(y))])

  # plot the decision surface

  x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
  x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
  xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),

  np.arange(x2_min, x2_max, resolution))

  Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
  Z = Z.reshape(xx1.shape)
  plt.contourf(xx1, xx2, Z, alpha=0.4, cmap=cmap)
  plt.xlim(xx1.min(), xx1.max())
  plt.ylim(xx2.min(), xx2.max())

  # plot class samples

  for idx, cl in enumerate(np.unique(y)):

    plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1],
    alpha=0.8, color=cmap(idx),
    marker=markers[idx], label=cl)

from sklearn.linear_model import Perceptron
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=30, n_features=2, n_informative=2, n_redundant=0, weights=[.3, .3], random_state=300)
plt.scatter(X[:,0], X[:,1], s=50)
pct = Perceptron(max_iter=100, verbose=0, random_state=300,
fit_intercept=True, eta0=0.002)
pct.fit(X, y)
plot_decision_regions(X, y, classifier=pct)
plt.title('Perceptron')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
